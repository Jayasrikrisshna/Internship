###  Q1 Most viewed Videos on YouTube


```python
#Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
#Rank, Name, Artist, Upload date, Views

import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

import time
from selenium.common.exceptions import NoSuchElementException


driver = webdriver.Chrome(r'C:\Users\Jayasri\Downloads\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome('chromedriver.exe')

url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
driver.maximize_window()
```

```python
rank=[]

try:
    rank_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rank_tag:
        rank.append(i.text.split('.')[0])
        
except NoSuchElementException:    
    rank.append('-')
except StaleElementReferenceException:
    rank.append('-')

name=[]

try:
    name_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in name_tag:
        name.append(i.text.split('"')[1])
        
except NoSuchElementException:    
    name.append('-')
except StaleElementReferenceException:
    name.append('-')

artist=[]

try:
    artist_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist_tag:
        artist.append(i.text)
        
except NoSuchElementException:    
    artist.append('-')
except StaleElementReferenceException:
    artist.append('-')

views=[]

try:
    view_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view_tag:
        views.append(i.text)
        
except NoSuchElementException:    
    views.append('-')
except StaleElementReferenceException:
    views.append('-')

    
date=[]

try:
    date_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date_tag:
        date.append(i.text)
        
except NoSuchElementException:    
    date.append('-')
except StaleElementReferenceException:
    date.append('-')

    
videos={'Rank':rank,'Name':name,'Artist':artist,
        'Views':views,'Date':date}
videos=pd.DataFrame(data=videos)
videos
    
```

### 5. The details of trending repositories on Github.com.

```python
driver = webdriver.Chrome(r'C:\Users\Jayasri\Downloads\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome('chromedriver.exe')

url='https://github.com/'
driver.get(url)
driver.maximize_window()
```

```python
exp_btn = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details")
exp_btn.click()
```

```python
trend_url = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
urls=trend_url.get_attribute("href")
driver.get(urls)
```

```python
# making empty lists    
repo_urls=[]
Contributors_count=[]
Language_used=[]
repo_title=[]
repo_desc=[]

#scraping repositories urls
repos=driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))
time.sleep(2) 

    
#scraping repositories title
repo_title=[]
try:
    repos=driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']")
    for i in repos:
        repo_title.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_title.append('No details available')
time.sleep(3)
    

for i in repo_urls:
    driver.get(i)
    l=[]
    time.sleep(3)
    #scraping repositories discription
    try:
        repo = driver.find_element(By.XPATH,"/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[1]/div/p")
        repo_desc.append(repo.text)
    except NoSuchElementException:
        repo_desc.append('-')
    
    #scraping contributors count
    try:
        count=driver.find_element(By.XPATH,"//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:#handling no such element exception
        Contributors_count.append('No details available')
    time.sleep(2)
    
     #scraping languages used   
    languages=driver.find_elements(By.XPATH,"//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            l.append(i.text)
    else:
        l.append('No languages used')
    Language_used.append(l)   
    
time.sleep(2)   
#creating dataframe
```

```python
df=pd.DataFrame({'Title':repo_title,
                'Description':repo_desc,
                'Contributors_count':Contributors_count,
                'Language_used':Language_used})
df
```
